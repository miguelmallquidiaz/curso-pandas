## nrows: Solamente mostrar las n primeras filas

```python
pd.read_csv('./data/titanic.csv', delimiter=';', nrows=3)
```

## usecols: Utilizar columnas especificas

```python
pd.read_csv(
    './data/titanic.csv', 
    delimiter=';', 
    usecols=["Survived", "Nombre", "Sexo"]
)
```

## Skiprows: Saltarse un número especifico de filas al inicio, se va a borrar los encabezados.

```python
pd.read_csv(
    './data/titanic.csv', 
    delimiter=';', 
    skiprows=3
)
```

## sheet_name = esto es para excel va a selecionar la pestaña en el libro de excel.

```python
pd.read_excel(
    './data/titanic.xlsx',
    sheet_name="Pasajeros_2")
```

## header = Se utiliza cuando las columnas no tienen encabezado
## names = Para darle nombre a las columnas deben coincidir con el número de columnas

```python
pd.read_excel(
    './data/titanic.xlsx',
    sheet_name="Pasajeros_2",
    header=None,
    names=["Id_Pasajero", "Survived", "Clase", "Nombre", "Sexo", "Edad", "Familiares", 
          "Tiquete", "Tarifa", "Cabina", "Embarcado"]
)
```

## Exportar archivos

Es importante indicarle la extensión

```python
df_titanic = pd.read_excel(
    './data/titanic.xlsx'
)
```

```python
# Exportar en Excel
# Si no queremos que pandas no genere una columna con indice
df_titanic.to_excel("titanic_export.xlsx", index=False)
```

```python
# Exportar en csv

df_titanic.to_csv("titanic_export.csv")
```

```python
# Exportar en txt
# Colocarlo por un separador
df_titanic.to_csv("titanic_export.txt", sep=";")
```

```python
# Exportar a Html
df_titanic.to_html("titanic_export.html")
```
