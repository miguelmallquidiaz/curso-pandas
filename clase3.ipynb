```python
# Seleccionar una serie o varias series

df[["Nombre", "Sexo"]]
```

```python
# Seleccionar los hombres que abordaron el titanic

men = df[df["Sexo"] == "Masculino"]
men.head()
```

```python
# Seleccionar las mujeres que abordaron el titanic

women = df[df["Sexo"] == "Femenino"]
women.head()
```

```python
# Persona con más de 30 años que abordan el titanic

age = df[df["Edad"] > 30]
age.head()
```

```python
# Contar cuantas personas tiene esa edad
age.shape
```

```python
# Personas que pertenecen a la clase 1 y 2 del titanic
people = df[df['Clase'].isin([1, 2])]
people.head()
```

```python
# Personas que pertenecen a la clase 1 o 2 del titanic
people = df[(df['Clase'] == 1) | (df['Clase'] == 2)]
people.head()
```

```python
# Obtener cuantos datos vacios hay
df.isnull().sum()
```

```python
# personas que no registraron su edad al abordar el barco
sin_edad = df[df["Edad"].isna()]
sin_edad.head(50)
```

```python
sin_edad.isnull().sum()
```

```python
# Personas que si registraron su edad
con_edad_registrada = df[df["Edad"].notna()]
con_edad_registrada.head(50)
```

```python
con_edad_registrada.isnull().sum()
```
