```python
## El objetivo es contar cuantas descripciones tiene la palabra tropical

data = {'description': [
    "This wine has a tropical and citrusy aroma.",  # Contiene "tropical"
    "A fruity and rich taste with hints of oak.",  # No contiene "tropical"
    "Bright acidity with tropical flavors of mango.",  # Contiene "tropical"
    "A dry wine with herbal notes."  # No contiene "tropical"
]}

reviews = pd.DataFrame(data)

# Explicación paso a paso
# .map() -> Recorrer cada fila es como un for
# lambda te permite escribir todo en una sola fila sintaxis lambda argumentos : operación como a + b
# desc es un argumento de lambda que representa cada fila de la columna en la que esta en operación
# in desc va a verificar y reemplaza el texto con true si la palabra buscada esta en esa fila y si no false
# sum() convierte True en 1 y False en 0 y luego suma todos los 1

n_trop = reviews.description.map(lambda desc: "tropical" in desc).sum()


print(n_trop)  # Output: 2
```

### Grouping and Sorting


```python
reviews= pd.read_csv('./data/winemag-data-130k-v2.csv')
reviews.head()
```

```python
# Agrupar el precio
# Se seleciona la columna points y se obtiene el valor máximo de ese points dentro de price
# Ordena los resultados por el indice que es el price

reviews.groupby('price').points.max().sort_index()
```

```python
# Obtener los precios minimos y maximos por cada variedad

# La función agg permite aplicar varias funciones de agregación
price_extremes = reviews.groupby('variety').price.agg([min, max])
price_extremes
```

```python
# ¿Cuáles son las variedades de vino más caras?

price_extremes.sort_values(by=['min', 'max'], ascending=False)
```

```python
# Crea una serie cuyo índice sea revisores y cuyos valores sean la puntuación media de las reseñas otorgadas por ese revisor.

reviewer_mean_ratings = reviews.groupby('taster_name').points.mean()
reviewer_mean_ratings
```

```python
# ¿Qué combinación de países y variedades son las más comunes?
# Ordene los valores de la serie en orden descendente según la cantidad de vinos.

country_variety_counts = reviews.groupby(['country', 'variety']).size().sort_values(ascending=False)
country_variety_counts
```

```python
## Crear una copia para no afectar el dataset original
reviews_copy = reviews.copy()
reviews_copy
```

```python
# What is the data type of the points column in the dataset?

dtype = reviews_copy['points'].dtype
dtype
```

```python
# Crea una serie a partir de las entradas de la columna de puntos, pero convierte las entradas en cadenas

point_strings = reviews_copy['points'].astype('str')
point_strings
```

```python
# obtener cuales son los tipos de datos de las columnas
reviews_copy.dtypes
```

```python
# A veces, la columna de precio es nula. ¿En cuántas reseñas del conjunto de datos falta un precio?

reviews_copy['price'].isnull().sum()
```

```python
# ¿Cuáles son las regiones productoras de vino más comunes?
# Este campo suele contener datos faltantes, por lo que debe reemplazar los valores faltantes por "Unknown"
# Ordene en orden descendente

# lo que hace fillna es reemplazar los valores faltantes
# value_counts() contar cuantas veces aparece el valor de forma única.
reviews_copy['region_1'].fillna('Unknown').value_counts().sort_values(ascending=False)
```

```python
'''
region_1 and region_2 are pretty uninformative names for locale columns in the dataset.
Create a copy of reviews with these columns renamed to region and locale, respectively.
'''

reviews_copy.rename(columns={'region_1':'region', 'region_2':'locale'})
```

```python
# Set the index name in the dataset to wines.

# renombra el indice de las filas 
reindexed = reviews_copy.rename_axis('wines', axis='rows')
reindexed
```
